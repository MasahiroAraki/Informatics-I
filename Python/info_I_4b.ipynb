本講座第1章のスライドから文章を抜き出したものを読み込み。

```
with open('chap1.txt') as f:
  s = f.read() 
s
```

単語分割

```
import spacy
nlp = spacy.load('ja_ginza')
```

```
doc = nlp(s)
textlist = [w.text for w in doc]
text = ' '.join(textlist)
text
```

表示対象としない単語をストップワードとしてリストアップ。

```
stop_words_ja = ['ある','する','できる','よう','ない','なる','こと','など','です','ます','その','とき','あれ','これ','いう']
```

ワードクラウドを表示

```
from wordcloud import WordCloud
import matplotlib.pyplot as plt
fontpath = '/usr/share/fonts/truetype/fonts-japanese-gothic.ttf'
def wc(word_chain):
  wordcloud = WordCloud(background_color="white",
                        font_path=fontpath,
                        max_words=30,
                        width=500,
                        height=200,
                        contour_width=1,
                        contour_color="black",
                        stopwords=stop_words_ja).generate(word_chain)
  plt.figure(figsize=(15,20))
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.show()
```

```
wc(text)
```
